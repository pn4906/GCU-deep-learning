```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
df = pd.read_csv('C:/datasets/abalone.csv')
df
```

```python
# 'Sex' 열을 One-Hot Encoding
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
```

```python
df.head()
```

```python
X = df.drop(['Rings', 'id'], axis=1)
y = df['Rings']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# LSTM용 데이터 형태 변환 (3차원 배열: [samples, timesteps, features])
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
lstm_model = models.Sequential([
    layers.LSTM(64, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # 회귀를 위한 출력 레이어
])

lstm_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
lstm_model.fit(X_train_lstm, y_train, epochs=10, validation_data=(X_test_lstm, y_test))
```

```python
lstm_test_loss, lstm_test_mae = lstm_model.evaluate(X_test_lstm, y_test)
```

```python
print(f"LSTM Test Loss: {lstm_test_loss}, Test MAE: {lstm_test_mae}")
```
